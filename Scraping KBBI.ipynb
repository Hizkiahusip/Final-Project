```python
import requests
import sqlite3
from sqlite3 import Error
import os
import re
from bs4 import BeautifulSoup
import sys, time
from itertools import cycle



def create_connection(db_file):
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)
    return None


def select_and_update(conn):
    i=1
    cur = conn.cursor()
    cur.execute("SELECT * FROM kata_berimbuhan")
    rows = cur.fetchall()

    for row in rows:
        print(str(row[0])+"/"+str(rows.__len__()))
        print(row)
        postag_kbbibud(row[2])

        # pt = postag_kbbiwebid(row[1])
        # cur.execute('UPDATE datakata SET postag =(?) WHERE _id=(?)',(pt,int(row[0])))
        # conn.commit()
        # os.system('cls||clear')


def postag(kata):
    response = requests.get("https://kbbi.kemdikbud.go.id/entri/"+kata)
    soup = BeautifulSoup(response.text, 'html.parser')
    results = soup.find_all('div', attrs={'class': 'container body-content'})
    result = results[0]
    ol = result.find('ol')
    try:
        return (ol.li.i.span.text)
    except:
        return ('')

def postag_kbbiwebid(kata):
    response = requests.get("https://kbbi.web.id/"+kata)
    soup = BeautifulSoup(response.text, 'html.parser')
    results = soup.find_all('div', attrs={'id': 'd1'})
    result = results[0]
    try:
        return result.find('em').text
    except:
        return ''

def postag_kbbiwebid_raw(kata):
    response = requests.get("https://kbbi.web.id/"+kata)
    soup = BeautifulSoup(response.text, 'html.parser')
    # results = soup.find_all('div', attrs={'id': 'd1'})
    results = soup.find_all('div', attrs={'id': 'd1'})
    result = results[0]
    try:
        print (result.find_all('div',attrs={'id': 'info'})[0].text)
    except:
        print ('nothing')



def postag_kbbibud(kata):
    response = requests.get("https://kbbi.kemdikbud.go.id/entri/"+kata)
    soup = BeautifulSoup(response.text, 'html.parser')
    results = soup.find_all('div', attrs={'class': 'container body-content'})
    result = results[0]
    try:
        if (result.find('ol')):
            ol = result.find('ol')
            print(ol.li.i.span.text)
        else:
            ul = result.find('ul')
            print(ul.li.i.span.text)
    except:
        print('')

def main():
    database = "C:\db\kbbi.db"

    conn = create_connection(database)
    with conn:
        select_and_update(conn)

if __name__ == '__main__':
    main()
```
