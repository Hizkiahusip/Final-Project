```python
import requests
import sqlite3
from sqlite3 import Error
import os
import re
import sys, time
import PyPDF2
from itertools import cycle

#membuat koneksi
def create_connection(db_file):
    try:
        conn = sqlite3.connect(db_file)
        return conn
    except Error as e:
        print(e)
    return None

def txt():
    isi = []
    # membuka file txt dengan membuat variabel nya jadi file 
    file = open("D:\\hizkia\\TA2\\implementasi\\banjir.TXT", 'rb')
    # membaca file
    data = file.read().decode('utf-8')
    #teks1 = re.sub('\n',' ',data)
    # menghapus tanda 	
    teks2 = re.sub('	','',data)
    teks = re.sub('	','',teks2)
    # menghapus angka
    result = ''.join(i for i in teks if not i.isdigit())
    # menghapus semua karakter ga berguna
    clean = result.translate({ord(i): None for i in '~`!@#$%^&*)(-_+=}{]\|:;"''<,>?/"“[”…••–'})
    # mengecilkan huruf
    lower = clean.lower()
    #split = lower.split(".")
    # kalau menemukan "." atau "\n" dijadikan satu array
    split = re.split('[.\n]',lower)
    # loop
    for j in range (46,len(split)):
        # batas yang diterima kaliamt
        if len(split[j]) > 14:
            k=0
            # menghapus spasi yang diawal
            while(split[j][k]==' ' or split[j][k]=='	'):
                split[j] = '' + split[j][k+1:]
            #menambahkan <s> dan </s> di awal dan akhir kalimat
            split[j] = '<s>' + split[j][0:] + '</s>'
            kalimat = split[j]
            #memasukkan nilai kalimat ke variabel isi.
            isi.append(kalimat)
    #                 cur = conn.cursor()
    #                 cur.execute('INSERT INTO korpus_kalimat (kalimat) VALUES (?)',(kalimat,))
    #                 conn.commit()
            print(kalimat)
    file.close()

def main():
    database = "C:\db\korpus.db"
    conn = create_connection(database)
#     with conn:
#         pdf(conn)
    txt()    
main()
```
